```python
# -------------------------------------------
# Project 6: Exploring Movie Data with Pandas
# Author: Riadh Moudir
# Date: 6 May 2025
# Description: This program analyzes a movie dataset using pandas.
# -------------------------------------------

import pandas as pd

# STEP 1: Load the CSV dataset
df = pd.read_csv("movie_dataset.csv")

# STEP 2: View the first few rows
print("🔹 First 5 movies:")
print(df.head())

# STEP 3: General info about the data
print("\n🔹 Dataset Info:")
print(df.info())

# STEP 4: Summary statistics
print("\n🔹 Summary Statistics:")
print(df.describe())

# STEP 5: Filter movies with rating > 8.0
print("\n🔹 Movies with rating > 8.0:")
print(df[df["Rating"] > 8.0])

# STEP 6: Filter movies with revenue > 500 million
print("\n🔹 Movies with revenue > 500M:")
print(df[df["Box Office (Millions)"] > 500])

# STEP 7: Sort by Release Year (newest first)
print("\n🔹 Movies sorted by Release Year:")
print(df.sort_values(by="Release Year", ascending=False))

# STEP 8: Sort by Revenue (highest first)
print("\n🔹 Movies sorted by Revenue:")
print(df.sort_values(by="Box Office (Millions)", ascending=False))

# STEP 9: Calculate average rating
avg_rating = df["Rating"].mean()
print(f"\n🔹 Average movie rating: {avg_rating}")

# STEP 10: Calculate total budget
total_budget = df["Budget (Millions)"].sum()
print(f"🔹 Total budget of all movies: ${total_budget}M")

# STEP 11: Movie with highest revenue
top_revenue_movie = df.loc[df["Box Office (Millions)"].idxmax()]
print("\n🔹 Movie with highest revenue:")
print(top_revenue_movie)

# STEP 12: Group by Genre (average rating)
print("\n🔹 Average Rating by Genre:")
print(df.groupby("Genre")["Rating"].mean().reset_index())

# STEP 13: Group by Director (total revenue)
print("\n🔹 Total Revenue by Director:")
print(df.groupby("Director")["Box Office (Millions)"].sum().reset_index())

# STEP 14: Add a new column for Profit
df["Profit (Millions)"] = df["Box Office (Millions)"] - df["Budget (Millions)"]

# STEP 15: Movie with highest profit
highest_profit = df.loc[df["Profit (Millions)"].idxmax()]
print("\n🔹 Movie with Highest Profit:")
print(highest_profit)
```
