```python
# ----------------------------------------
# Project: Simple Banking System
# Author: Riadh Moudir
# Date: 6 May 2025
# Description: A console-based banking system with login, balance management, and menu options.
# ----------------------------------------

# -------- 1. USER LOGIN (up to 3 attempts) --------
# Hardcoded username and password
correct_username = "user123"
correct_password = "pass123"

# Allow the user to try login 3 times
login_attempts = 0
logged_in = False

while login_attempts < 3:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == correct_username and password == correct_password:
        print("‚úÖ Login successful!\n")
        logged_in = True
        break
    else:
        print("‚ùå Incorrect username or password. Try again.\n")
        login_attempts += 1

if not logged_in:
    print("üö´ Too many failed attempts. Exiting system.")
else:
    # -------- 2. MAIN MENU --------
    balance = 1000  # Starting balance

    while True:
        print("\n=== Banking System Menu ===")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        # -------- 3. BALANCE MANAGEMENT --------

        if choice == "1":
            # CHECK BALANCE
            print(f"üí∞ Your current balance is: ${balance}")

        elif choice == "2":
            # DEPOSIT MONEY
            try:
                amount = float(input("Enter amount to deposit: "))
                if amount > 0:
                    balance += amount
                    print(f"‚úÖ Deposited ${amount}. New balance: ${balance}")
                else:
                    print("‚ùå Please enter a positive amount.")
            except ValueError:
                print("‚ùå Invalid input. Please enter a number.")

        elif choice == "3":
            # WITHDRAW MONEY
            try:
                amount = float(input("Enter amount to withdraw: "))
                if amount <= 0:
                    print("‚ùå Amount must be positive.")
                elif amount > balance:
                    print("‚ùå Not enough balance.")
                else:
                    balance -= amount
                    print(f"‚úÖ Withdrawn ${amount}. New balance: ${balance}")
            except ValueError:
                print("‚ùå Invalid input. Please enter a number.")

        elif choice == "4":
            # EXIT SYSTEM
            print("üëã Thank you for using our banking system.")
            break

        else:
            print("‚ùå Invalid option. Please choose 1, 2, 3, or 4.")
```
