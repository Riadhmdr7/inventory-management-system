```python
# ----------------------------------------
# Project 3: Movie Library System
# Author: Riadh Moudir
# Date: 6 May 2025
# Description: This program lets the user manage a list of movies using a console interface.
# ----------------------------------------

# Create a list to hold the movies
movie_library = []

# Function to add a new movie
def add_movie():
    title = input("Enter movie title: ")
    genre = input("Enter movie genre: ")
    year = input("Enter release year: ")
    rating = input("Enter movie rating (e.g. 7.5): ")

    # Create a dictionary to store movie info
    movie = {
        "title": title,
        "genre": genre,
        "year": year,
        "rating": rating
    }

    movie_library.append(movie)
    print(f"‚úÖ Movie '{title}' added successfully!")

# Function to display all movies
def view_movies():
    if not movie_library:
        print("üì≠ No movies in the library yet.")
    else:
        print("\nüéûÔ∏è Movie Library:")
        for i, movie in enumerate(movie_library, start=1):
            print(f"{i}. {movie['title']} ({movie['year']}) - {movie['genre']} - Rating: {movie['rating']}")

# Function to search for a movie by title
def search_movie():
    keyword = input("Enter title to search: ").lower()
    found = False
    for movie in movie_library:
        if keyword in movie['title'].lower():
            print(f"üîç Found: {movie['title']} ({movie['year']}) - {movie['genre']} - Rating: {movie['rating']}")
            found = True
    if not found:
        print("‚ùå No movie found with that title.")

# Main menu loop
while True:
    print("\n=== Movie Library Menu ===")
    print("1. Add Movie")
    print("2. View Movies")
    print("3. Search Movie")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        add_movie()
    elif choice == "2":
        view_movies()
    elif choice == "3":
        search_movie()
    elif choice == "4":
        print("üëã Goodbye! Thanks for using the Movie Library System.")
        break
    else:
        print("‚ùå Invalid choice. Please enter 1-4.")
```
