```python
# ----------------------------------------
# Project: Introduction to Python - Basic Commands
# Author: Riadh Moudir
# Date: 6 May 2025
# Description: This project demonstrates basic Python features including variables,
# data types, input/output, conditionals, loops, operators, and more.
# ----------------------------------------

# -------- Task 1: Assigning Values to Variables --------
# Create one int, one float, and one string
my_int = 7
my_float = 3.14
my_string = "Hello"
print(my_int, my_float, my_string)

# -------- Task 2: Data Types --------
age = 18
print("Data type of age:", type(age))  # Should show <class 'int'>
age = 18.5
print("Data type after changing age to float:", type(age))

is_student = True
print("Data type of is_student:", type(is_student))  # <class 'bool'>

# -------- Task 3: Basic Input and Output --------
name = input("What is your name? ")
print("Hello,", name + "!")

user_age = int(input("Enter your age: "))
if user_age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")

# -------- Task 4: Converting Data Types --------
num_str = input("Enter a number (as string): ")
num_int = int(num_str)
print("Square of your number:", num_int ** 2)

float_str = "4.56"
converted_float = float(float_str)
print("Float * 2:", converted_float * 2)

# -------- Task 5: Arithmetic Operators --------
print("45 + 67 =", 45 + 67)
print("156 - 89 =", 156 - 89)
print("12 * 7 =", 12 * 7)
print("100 / 4 =", 100 / 4)
print("10 % 3 =", 10 % 3)  # Remainder
print("3 ** 4 =", 3 ** 4)  # 3^4

# -------- Task 6: Comparison Operators --------
a = 10
b = 20
print("a == b:", a == b)
print("a != b:", a != b)
print("a < b:", a < b)
print("a > b:", a > b)
print("a <= b:", a <= b)
print("a >= b:", a >= b)

# Comparison with user input
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
if x >= y:
    print("First number is greater than or equal to second.")
else:
    print("First number is less than second.")

# -------- Task 7: Logical Operators --------
number = int(input("Enter a number between 10 and 20: "))
if number > 10 and number < 20:
    print("Your number is valid (between 10 and 20).")
else:
    print("Number is out of range.")

num2 = int(input("Enter a number to test for <5 or >50: "))
if num2 < 5 or num2 > 50:
    print("The number is either less than 5 or greater than 50.")
else:
    print("The number is between 5 and 50.")

# -------- Task 8: Conditional Statements --------
user_age = int(input("Enter your age again: "))
if user_age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

n = int(input("Enter a number to test for FizzBuzz: "))
if n % 3 == 0 and n % 5 == 0:
    print("FizzBuzz")
elif n % 3 == 0:
    print("Fizz")
elif n % 5 == 0:
    print("Buzz")
else:
    print(n)

# -------- Task 9: Loops --------
print("For loop from 1 to 10:")
for i in range(1, 11):
    print(i, end=' ')
print()

print("Even numbers from 2 to 20 (while loop):")
even = 2
while even <= 20:
    print(even, end=' ')
    even += 2
print()

n = int(input("Enter a number to show values divisible by 4: "))
print("Numbers divisible by 4 from 1 to", n)
for i in range(1, n + 1):
    if i % 4 == 0:
        print(i, end=' ')
print()

# -------- Bonus Task: Nested Loops (Pattern Printing) --------
print("Triangle pattern using nested loops:")
for i in range(1, 6):  # Outer loop for rows
    for j in range(1, i + 1):  # Inner loop for columns
        print("*", end="")
    print()  # Move to next line
```
